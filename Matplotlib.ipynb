```python
a =[1,2,3,4,5]
b =[5,6,99,8,9]

plt.figure(figsize=(4,4))
plt.plot(a,b)
plt.title("Random")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
```
